##Continuity and Differentiaition

Limits are important for understanding continuity.
\begin{align}
\lim_{x→a}f(a) = L
\end{align}

Describes a continuous function where there arent any abrupt "turns, cusps, and\or jumps." With this definition of continuity with a limit, we have the derrivative as:

\begin{align}
f'(a) = \lim_{h→∞} \frac{f(a+h) - f(a)}{h}
\end{align}

One important theorm we need is the Mean Value Theorem:
\begin{align}
\text{Let  } f:[a,b]→\mathbf{R}\text{ where f is continuous and differantiable on (a,b)}\\
\text{then there exists a, }c∈(a,b) s.t.f'(c)=\frac{f(b)-f(a)}{b-a}
\end{align}

##Taylor's Theorem

Let $f:D→\mathbf{R}$ where $D⊆\mathbf{R}$. Suppose $f$ has a $m$
times continuous derivative on $[a,b]$. Then
\begin{align}
f(b)=f(a)+(b-a)f'(a)+\frac{1}{2}(b-a)^2f''(a)+\dots+\frac{(b-a)^{m-1}}{(m-1)!}f^{(m)}(a)+R_M
\end{align}
where $R_M = \frac{(b-a)^m}{(m)!}f^{(m)}(a+θ(b-a))$  for some $0<\theta<1$.

As you can see this is a generalizaiton of the Mean Value Theorem. We can use numpy to find the taylor approximations of diffrtent functions.

In this demonstration we will be showing how much error our apporximation of the $cos(x)$ function has with diffrent numbers of coefficients.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-3.14, 3.14, 100)

# Compute the function values
y = np.cos(x)

# Perform linear regression to obtain coefficients for a polynomial of degree 3
coefficients = np.polyfit(x, y, 3)

# Generate the Taylor expansion using the coefficients
taylor_expand = np.polyval(coefficients, x)

# Plot the original function and its Taylor expansion
plt.plot(x, y, label='cos(x)')
plt.plot(x, taylor_expand, label='Taylor Expansion', linestyle='--')
plt.title('Taylor Expansion')
plt.legend()
plt.show()
```

```python
error_values = []
for i in range(20):
  # Perform linear regression to obtain coefficients for a polynomial of degree i
  coefficients = np.polyfit(x, y, i+1)

  # Generate the Taylor expansion using the coefficients
  taylor_expand = np.polyval(coefficients, x)

  #Determine error between Taylor expansion and function
  error = np.abs(taylor_expand - y)
  error_values.append(error)

for i in range(len(error_values[0:4])):
  plt.plot(x, error_values[i]);

leg = [str(i+1)+" Taylor Coefficients" for i in range(len(error_values))]
plt.legend(leg)
plt.title("Error of Taylor Expansion for diffrent number of coefficients")
plt.show();
```

```python
for i in range(len(error_values[4::])):
  plt.plot(x, error_values[i+4]);

leg = [str(i+4)+" Taylor Coefficients" for i in range(len(error_values[4::]))]
plt.legend(leg)
plt.title("Error of Taylor Expansion for diffrent number of coefficients")
plt.show();
```

```python
for i in range(len(error_values[9::])):
  plt.plot(x, error_values[i+9]);

leg = [str(i+9)+" Taylor Coefficients" for i in range(len(error_values[9::]))]
plt.legend(leg)
plt.title("Error of Taylor Expansion for diffrent number of coefficients")
plt.show();
```

Notice how the error gets increasingly small in our range.
